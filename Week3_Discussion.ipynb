Load and rotate an image


```
import scipy
from scipy import misc
import numpy as np
import imageio
from matplotlib import pyplot as plt
from time import time
from skimage.data import shepp_logan_phantom

img = shepp_logan_phantom()

plt.figure()
plt.imshow(img, cmap='gray')

print('The shape of the phantom image is: {} '.format(np.shape(img)))
```

Rotate the image for 90 degrees.

```
from skimage.transform import rotate

rotated_image = rotate(img, 90)
plt.figure()
plt.imshow(rotated_image, cmap='gray')
plt.show()
```

### Performing Radon Transform using Scikit-image Package 

```
from skimage.transform import radon
# full spectrum radon transform
num_projections = len(img) # number of pixels, to get # cols of 1D proj
print(num_projections) # 400

# divide space from 0-180 linearly given 400 pts

theta = np.linspace(0, 180.0, num_projections)
sinogram = radon(img, theta)

plt.figure()
plt.imshow(sinogram, cmap='gray')
print("Radon transform using scikit image")
plt.show()
```
